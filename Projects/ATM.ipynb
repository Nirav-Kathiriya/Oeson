```python
# Empty dict for user details
users = {}

while True:
    des = '*' * 70
    print(des)
    print("\tWelcome to the ATM System created by Nirav Kathiriya")
    print(des)

    # Asking for new or old user
    ask_user = input('\nEnter 1 for New User or 0 for Existing User: ')

    if ask_user == '1':
        # Taking details of the New User
        f_name = input('Enter First Name: ')
        l_name = input('Enter Last Name: ')
        
        # uid is unique or not
        while True:
            uid = input('Enter User ID: ')
            if uid in users:
                print("User ID is already taken. Please enter a different User ID.")
            else:
                break
    
        while True:
            pin = input('Enter Your PIN: ')
            confirm_pin = input('Confirm Your PIN: ')
            if pin == confirm_pin:
                break
            print('\nPINs do not match. Please try again.')

        try:
            opening_balance = int(input('Enter Opening Balance: '))
        except ValueError:
            print("\nInvalid amount! Please enter a valid number.")
            continue
        
        users[uid] = {'pin': pin, 'balance': opening_balance}
        print(f'\nUser {f_name} {l_name} registered successfully!\n')

    elif ask_user == '0':
        # Checking if the user is new or old
        check_uid = input('Enter Your User ID: ')
        check_pin = input('Enter Your PIN: ')
        
        # Allow up to 3 attempts only
        attempts = 0
        while attempts < 3:
            if check_uid in users and users[check_uid]["pin"] == check_pin:
                balance = users[check_uid]["balance"]
                print(f"\nWelcome, {check_uid}!\n")
                
                while True:
                    print("\tEnter 1 to Check Balance")
                    print("\tEnter 2 to Withdraw Money")
                    print("\tEnter 3 to Deposit Money")
                    print("\tEnter 9 to Exit")

                    try:
                        reply = int(input('\nEnter Your Choice: '))
                    except ValueError:
                        print('\nInvalid input! Please enter a valid number (1, 2, 3, or 9).')
                        continue

                    if reply == 1:
                        print(f"\nYour Current Balance is: £{balance}\n")

                    elif reply == 2:
                        print(f"\nYour Current Balance is: £{balance}")
                        try:
                            withdraw = int(input('\nEnter Amount to Withdraw: '))
                            if withdraw <= 0:
                                print('\nWithdrawal amount must be greater than zero.')
                                continue
                            if withdraw <= balance:
                                balance -= withdraw
                                users[check_uid]["balance"] = balance
                                print(f'\nWithdrawal successful! New balance = £{balance}\n')
                            else:
                                print('\nInsufficient balance. Please enter a valid amount.\n')
                        except ValueError:
                            print('\nInvalid withdrawal amount! Please enter a valid number.\n')

                    elif reply == 3:
                        try:
                            deposit = int(input('\nEnter Amount to Deposit: '))
                            if deposit <= 0:
                                print('\nDeposit amount must be greater than zero.')
                                continue
                            balance += deposit
                            users[check_uid]["balance"] = balance
                            print(f'\nDeposit successful! New balance = £{balance}\n')
                        except ValueError:
                            print('\nInvalid deposit amount! Please enter a valid number.\n')

                    elif reply == 9:
                        print('\nThank you for using the ATM service by Nirav Kathiriya!')
                        break
                    
                    else:
                        print('\nInvalid choice! Please select a valid option from 1, 2, 3, or 9 only.\n')

                continue_program = input('Do you want to perform another transaction? (Yes/No): ').lower()
                if continue_program == 'yes' or continue_program == 'no':
                    if continue_program != 'yes':
                        break
                else:
                    print('\n Please type Yes or No only.\n')
            
            else:
                attempts += 1
                if attempts < 3:
                    print(f"\nIncorrect username or PIN. You have {3 - attempts} attempts left.\n")
                    check_uid = input('Enter Your User ID: ')
                    check_pin = input('Enter Your PIN: ')
                else:
                    print('\nToo many incorrect attempts. Please try again later.\n')
                    break
    
    else:
        print("\nInvalid input! Please enter 1 for New User or 0 for Existing User.\n")
```
